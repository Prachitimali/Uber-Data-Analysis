```python
import pandas as pd
import numpy as nm 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading the dataset
data = pd.read_csv("UberDataset.csv")
data
```

```python
#cloning the dataset
clone_data = data.copy()
```

```python
# Check for missing values
print(clone_data.isnull().sum())
```

```python
# Handle missing values 
clone_data.fillna(0, inplace=True)  
```

```python
# Check data types
print(clone_data.dtypes)
clone_data
```

```python
# Check for duplicates
duplicates = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

# Remove duplicates
data.drop_duplicates(inplace=True)
```

```python
# Convert MILES to numeric
clone_data['MILES'] = pd.to_numeric(clone_data['MILES'], errors='coerce')

# Display updated types
print(clone_data.dtypes)
```

```python
# Remove outliers using the IQR method
numerical_columns = clone_data.select_dtypes(include=['int64', 'float64']).columns
for column in numerical_columns:
    Q1 = clone_data[column].quantile(0.25)
    Q3 = clone_data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter out outliers
    clone_data = clone_data[(clone_data[column] >= lower_bound) & (clone_data[column] <= upper_bound)]
```
